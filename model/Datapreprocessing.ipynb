### Preprocessing Food Data

```python
columns_list = food_data.columns.tolist()
columns_list
```

#### Conversion of all values to mg

```python
def convert_to_mg(value):
    if isinstance(value, str):
        try:
            if 'g' in value:
                return float(value.replace('g', '')) * 1000
            elif 'IU' in value:
                return float(value.replace('IU', '')) * 0.000025
            elif 'mg' in value:
                return float(value.replace('mg', ''))
            else:
                return float(value)
        except ValueError:
            return value
    return value
```

```python
# Exclude food_name column
columns_to_convert = food_data.columns[1:] 
food_data[columns_to_convert] = food_data[columns_to_convert].applymap(convert_to_mg)
```

```python
def mcg_to_mg(value):
    if isinstance(value, str):
        try:
            if 'mcg' in value:
                return float(value.replace('mcg', '')) * 0.001
            else:
                return float(value)
        except ValueError:
            return value
    return value

# value = '990mcg'
# result = mcg_to_mg(value)
# print(result)
```

```python
# Exclude food_name column
columns_to_convert = food_data.columns[1:] 
food_data[columns_to_convert] = food_data[columns_to_convert].applymap(mcg_to_mg)
```

```python
food_data.head(5)
```

```python
def remove_units(value):
    if isinstance(value, str):
        return value.replace('mg', '').replace('mcg', '').replace('IU', '').replace('g', '').strip()
    return value
```

```python
# Exclude food_name column
columns_to_convert = food_data.columns[1:] 
food_data[columns_to_convert] = food_data[columns_to_convert].applymap(remove_units)
```

#### Excluding some food that matches certain strings

```python
# ~ denotes case-insensitive manner, regardless of the case or capitalization
# Remove rows with food_name that matches with term 'babyfood, beverage'
food_data = food_data[~food_data['name'].str.contains('babyfood|beverage', case=False)]
```

```python
# Float conversion exclude name
columns_to_convert = food_data.columns.difference(['name'])
food_data[columns_to_convert] = food_data[columns_to_convert].astype(float)

print(food_data.dtypes)
```

#### Drop term 'raw' in name column

```python
import re

# Function to remove comma in front of 'raw' from a string
def clean_name(s):
    # Remove the comma in front of 'raw'
    cleaned_string = re.sub(r',\s*raw\b', '', s, flags=re.IGNORECASE).strip()
    return cleaned_string

# Apply the function to the 'name' column
food_data['name'] = food_data['name'].apply(clean_name)

# Drop duplicates based on the 'name' column
food_data = food_data.drop_duplicates(subset='name', keep='first')
```

#### Export Food Dataframe

```python
# Export the DataFrame to a CSV file
food_data.to_csv('preprocessed_food.csv', index=False)
```
